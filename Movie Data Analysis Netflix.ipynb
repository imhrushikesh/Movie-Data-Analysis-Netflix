```python
def catigorize_col(df, col, labels): 
    # Directly calculate the edges mathematically without using .describe()
    edges = [
        df[col].min(), 
        df[col].quantile(0.50), # Median
        df[col].quantile(0.75), 
        df[col].max()
    ] 
    
    # Segment the data into the bins
    df[col] = pd.cut(df[col], bins=edges, labels=labels, duplicates='drop', include_lowest=True) 
    return df

# Define your 3 labels to match the 3 bins created by your 4 edges
labels = ['not_popular', 'below_avg', 'popular']

# Run the function
catigorize_col(df, 'Vote_Average', labels)
```
